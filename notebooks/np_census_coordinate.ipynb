```python
map_ = gpd.read_file("../raw_data/nl/nl_ageb_urbana.shp")
```

```python
map_urban = map_.drop(columns=["CVE_AGEB","CVE_ENT","CVE_LOC","CVE_MUN"])
```

```python
map_urban.shape
```

```python
data = pd.read_csv("../raw_data/pob1.csv")
```

```python
data.head()
```

```python
data_ageb = data[data['NOM_LOC'] == "Total AGEB urbana"]
```

```python
data_ageb['CVEGEO'] = (data_ageb.ENTIDAD.astype(str).str.zfill(2) +\
                      data_ageb.MUN.astype(str).str.zfill(3) +\
                      data_ageb.LOC.astype(str).str.zfill(4) +\
                      data_ageb.AGEB.astype(str).str.zfill(4) \
                      )
```

```python
data_ageb.head()
```

```python
data_position = data_ageb.merge(map_urban, on="CVEGEO")
```

```python
data_position.head()
```

```python
data_position.to_csv("../raw_data/data_position.csv")
```
